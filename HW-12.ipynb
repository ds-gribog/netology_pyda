# Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

1)если источник traffic_source равен yandex или google, то в source_type ставится organic

2)для источников paid и email из России - ставим ad

3)для источников paid и email не из России - ставим other

4)все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
log_ = pd.read_csv('visit_log.csv', sep=';')

def source_type(row):
    if row['traffic_source'] in ['yandex', 'google']:
        return 'organic'
    elif row['traffic_source'] in ['paid', 'email']:
        return 'ad' if row['region'] == 'Russia' else 'other'
    else: 
        return row['traffic_source']

log_['source_type'] = log_.apply(source_type, axis = 1)

log_.tail(8)
```

# Задание 3
В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
from pymystem3 import Mystem
```

```python
%%time
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})

m = Mystem()

def lemms(row):
        lemmas = m.lemmatize(row['keyword'])
        return str(''.join(lemmas).strip())
        
data['lemmas'] = data.apply(lemms, axis=1)
data
```
