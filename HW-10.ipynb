# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

ratings = pd.read_csv('New/ratings.csv')
movies = pd.read_csv('New/movies.csv')

data = ratings[ratings.rating ==5].groupby('movieId').count().sort_values(by = ['rating'],ascending =[False]).head(1)

data.merge(movies, on = 'movieId')[['title', 'rating']]
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')

data[data['country'].isin(['Lithuania','Latvia','Estonia'])
     & data['category'].isin([4,12,21]) 
     & data['year'].isin(range(2005,2011)) 
     & data['quantity'] > 0].groupby('country').sum()[['quantity']]
```
